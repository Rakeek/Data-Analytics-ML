## Feature Selection

### LinearRegression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
for column in['Name','Genres','Type','Language']: # instead of writing many lines for each columns we can use a for loop to do it in one line.
    df[column] = le.fit_transform(df[column])
```

```python
df
```

```python
x=df[['Name','Genres','Type','Language','Total Seasons','Total Episodes']]
x
```

```python
y=df[['Rating']]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
model
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
y_test
```

```python
mse=mean_squared_error(y_test,y_predict)
print(mse)
```

```python
mae=mean_absolute_error(y_test,y_predict)
mae
```

```python
lin_r2score=r2_score(y_test,y_predict)
print(lin_r2score)
```

```python
new_experience=np.array([[51,3,51,21,21,1]])
predicted_rating=model.predict(new_experience)
print(f'predicted rating(0-10) {predicted_rating[0]}')
```
